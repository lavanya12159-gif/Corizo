```python
import kagglehub
import os
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

# Reload the dataset to define 'df'
path_to_dataset_dir = kagglehub.dataset_download("saurabh00007/iriscsv")
csv_file_path = os.path.join(path_to_dataset_dir, "Iris.csv")
df = pd.read_csv(csv_file_path)

# Prepare data for training
x=df.iloc[:,1:5]
y=df[["Species"]]
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8, random_state=42) # Added random_state for reproducibility

# Train Decision Tree Classifier
dtc = DecisionTreeClassifier(random_state=42)
dtc.fit(xtrain, ytrain.values.ravel())
print("Decision Tree Classifier trained successfully.")
```

```python
from sklearn.metrics import accuracy_score
y_pred_dtc = dtc.predict(xtest)
accuracy_dtc = accuracy_score(ytest, y_pred_dtc)
print(f"Decision Tree Classifier Accuracy: {accuracy_dtc:.2%}")
```

```python
from sklearn.metrics import confusion_matrix
conf_matrix_dtc = confusion_matrix(ytest, y_pred_dtc)
print("\nDecision Tree Classifier Confusion Matrix:")
print(conf_matrix_dtc)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(conf_matrix_dtc, annot=True, xticklabels=dtc.classes_, yticklabels=dtc.classes_, cmap='Blues')
plt.title('Confusion Matrix for Decision Tree Classifier')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state=42)
rfc.fit(xtrain, ytrain.values.ravel())
print("Random Forest Classifier trained successfully.")
```

```python
from sklearn.metrics import accuracy_score
y_pred_rfc = rfc.predict(xtest)
accuracy_rfc = accuracy_score(ytest, y_pred_rfc)
print(f"Random Forest Classifier Accuracy: {accuracy_rfc:.2%}")
```

```python
from sklearn.metrics import confusion_matrix
conf_matrix_rfc = confusion_matrix(ytest, y_pred_rfc)
print("\nRandom Forest Classifier Confusion Matrix:")
print(conf_matrix_rfc)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(conf_matrix_rfc, annot=True, xticklabels=rfc.classes_, yticklabels=rfc.classes_, cmap='Blues')
plt.title('Confusion Matrix for Random Forest Classifier')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```
